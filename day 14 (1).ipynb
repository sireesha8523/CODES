```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
taxi_df=sns.load_dataset('taxis')
taxi_df
```

```python
sns.barplot(x='payment',y='total',data=taxi_df,color='pink',estimator='std')
plt.show()
plt.tight_layout()
```

```python
sns.histplot(x=taxi_df['fare'],color='purple')
plt.show()
```

```python
sns.histplot(x=taxi_df['fare'],palette=['blue','black'],hue=taxi_df['payment'])
plt.show()
```

```python
sns.distplot(x=taxi_df['fare'],hist=True,kde=False)
plt.show()
```

```python
sns.boxplot(x=taxi_df['fare'],color='gold')
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
iris_data=sns.load_dataset('iris')
iris_data
```

```python
sns.scatterplot(x=iris_data['sepal_length'],y=iris_data['petal_length'])
plt.show()
```

```python
sns.stripplot(x=iris_data['species'],y=iris_data['sepal_width'],palette=['r','g','y'])
plt.show()
```
